```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('dataset/loans.csv').sample(10)
display(data.sample(10))
display(data.shape)
# grade: 贷款级别
# sub_grade: 贷款细分级别
# short_emp: 一年以内短期雇佣
# emp_length_num:　受雇年限
# home_ownership:居住状态（自有，按揭，租住）
# dti：贷款占收入比例
# purpose:贷款用途
# term:贷款周期
# last_delinq_none:贷款申请人是否有不良记录　
# last_major_derog_none:贷款申请人是否有还款逾期90天以上记录
# reｖol_util：透支额度占信用比例
# total_rec_late_fee:逾期罚款总额
# safe_loans:贷款是否安全

```

```python
from sklearn.preprocessing import LabelEncoder
from collections import defaultdict
```

```python
# 将以上非数值数据映射为数值型。
# 但是我们这里没有解决的一个问题是，每个不同的数都映射为一个数字，那么如果有很多很多小数的话
# 这里就爆炸了。理论上这里应该在网络结构中采用比较数值大小的方式。本例子适用于简单特征01型的决策
y = data.safe_loans
d = defaultdict(LabelEncoder)
X_trans = data.apply(lambda x : d[x.name].fit_transform(x))
y_train = y
display(X_trans.sample(10))
display(len(X_trans))
```

```python
from collections import  Counter
from tqdm import tqdm
```

```python

# ************ 计算信息熵 **************
yes = np.asarray(X_trans[X_trans[X_trans.columns[-1]]==1])
no = np.asarray(X_trans[X_trans[X_trans.columns[-1]]==0])
P_yes = len(yes)/len(X_trans)
P_no = len(no)/len(X_trans)
HX = - P_yes*np.log2(P_yes)- P_no*np.log2(P_no)
display("信息熵 = " +str(HX))

# ************ 计算条件熵 **************
# H存放条件熵
Gda = []
# 遍历每一特征列。除了标签列
for i in X_trans.columns[:-1]:
    # 存放条件熵
    Hi = 0
    # 获取当前特征每种情况出现的次数，以便计算每个情况各自的概率
    condProbCollections = Counter(X_trans[i]).items()
    # 每种特征可能有N中情况，累加
    for k in condProbCollections:
        # 获取当前条件X发生的总样本(包含所有列)
        samples_of_current = X_trans[X_trans[i]==k[0]]
        # 获取当前条件X发生的总样本(仅包含当前特征列)
        samples_of_current_features = samples_of_current[i]
        # 获取当前条件X发生下，被判定为安全和不安全的总次数
        total = len(samples_of_current_features)
        # 安全总次数
        k_safe = len(samples_of_current[samples_of_current["safe_loans"]==1])
        # 不安全总次数
        k_unsafe = total - k_safe
        # 计算安全和不安全的概率
        P_k_safe = k_safe/total
        P_k_unsafe = k_unsafe/total
        # 累加条件熵
        log_P_k_safe = 0 if P_k_safe==0 else np.log2(P_k_safe) # 防止出现0值报错
        log_P_k_unsafe = 0 if P_k_unsafe==0 else np.log2(P_k_unsafe)
        Hi +=  - (total/len(X_trans))*(P_k_safe * log_P_k_safe + P_k_unsafe * log_P_k_unsafe)
    # 保存信息增益
    Gda.append({"value":HX - Hi, "feature":i})

print("信息增益为")
print(Gda)
```

```python
# 构建决策树结构。决策树信息存储在JSON字典当中
def create_tree(Gda, node=False):
    # 递归出口
    if len(Gda)==0:
        return node.iloc[:1,-1].values[0]
    # 获取第一个特征列
    feature = Gda[0]['feature']
    # 删除该列，以便递归时不会重复到达这里
    del Gda[0]
    # 获取当前特征每种情况出现的次数，以便计算每个情况各自的概率
    condProbCollections = Counter(X_trans[feature]).items()
    # 定义树字典。必须使用特征feature作为键名
    tree = {feature:{}}
    for [value,counts] in condProbCollections:
        tree[feature][value] = create_tree(Gda, X_trans[X_trans[feature]==value])
    return tree
```

```python
# 给信息增益从大到小排序
a = np.asarray(Gda)
a = sorted(a,key=lambda value:value["value"], reverse=True)
print(a)


# 创建树
tree = create_tree(a)
tree
```
